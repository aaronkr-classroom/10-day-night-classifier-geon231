# Day and Night Image Classifier
---

The day/night image dataset consists of 200 RGB color images in two categories: day and night. There are equal numbers of each example: 100 day images and 100 night images.

We'd like to build a classifier that can accurately label these images as day or night, and that relies on finding distinguishing features between the two types of images!

*Note: All images come from the [AMOS dataset](http://cs.uky.edu/~jacobs/datasets/amos/) (Archive of Many Outdoor Scenes).*

## Load the datasets

````python
import matplotlib.pyplot as plt
import random

#config
num_img_per_class = 5
class_labels = ['day', 'night']

class1_img = [img for img, label in IMAGE_LIST if label ==class_labels[0]]
class2_img = [img for img, label in IMAGE_LIST if label ==class_labels[1]]

selected_class1 = random.sample(class1_img, num_img_per_class)
selected_class2 = random.sample(class2_img, num_img_per_class)

fig, axes = plt.subplots(2, num_img_per_class, figsize=(20, 10))
fig.suptitle("Day and Night images", fontsize=20)
for i in range(num_img_per_class):
    axes[0, i].imshow(selected_class1[i])
    axes[0, i].set_title("Day images")
    axes[0, i].axis('off')
    
    axes[1, i].imshow(selected_class2[i])
    axes[1, i].set_title("Night images")
    axes[1, i].axis('off')

plt.show()
````
